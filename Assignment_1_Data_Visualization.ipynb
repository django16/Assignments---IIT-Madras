# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```


#Loading the dataset


```python
Graph = pd.read_csv(r'/content/cars-1.csv')
Graph.head()
```

```python
plt.plot(Graph['model'],Graph['hp'])
plt.ylabel('HorsePower of cars')
plt.xlabel('Model Names of car')
plt.xticks(rotation='90')
plt.title("Model Names v/s HorsePower")
plt.show()
```

#Bar - PLot Graph

```python
plt.bar(x=Graph['carb'],height=Graph['gear'])
plt.ylabel('Number of Forward Gears')
plt.xlabel('Number of Carburetors')
plt.xticks(rotation='0')
plt.title("Carbs v/s Gear")
plt.show()
```

# Histogram

```python
plt.hist(Graph['wt'],bins=30)
plt.ylabel('Count')
plt.xlabel('weight of the cars')
plt.title("Histogram for the weight values")
plt.show()
```

```python
sns.histplot(x = Graph['wt'],kde=True).set(title='Histogram for the weight values')
```

# Pie - Chart

```python
fig = plt.figure(figsize= (15,40))
plt.pie(Graph['cyl'],labels=Graph['model'])
plt.show()
```

# Area Plot

```python
plt.fill_between(Graph['am'],Graph['carb'])
plt.ylabel('Number of carburetors')
plt.xlabel('Transmission')
plt.title("Transmission vs Number of carburetors")
plt.show()
```

```python
Graph.plot.area(x='am',y='carb')
```
